Image histogram visualizations per channel

```python
# Importing libraries
import tensorflow 
import keras  
import os
import glob 
from skimage import io 
import random 
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline

#Opening the images

#Name of the chosen pictures
image_names =["Countryside.jpg", "Bruce.jpg", "Wendland.jpg", "Bruxinha.jpg", "Bird.jpg"]
 #["Penguin.jpg", "Smiley.png", "Square.jpg", "Volkswagen.png", "Wifi.jpg"]

five_images = []
for j in range(5):
  five_images.append(io.imread(image_names[j]).astype('float32'))

#Dictionary for plotting the different colors
color_dictionary = {
  0: "Red",
  1: "Green",
  2: "BLue"
}


for image in five_images:

  

  for i in range(3): #Visualizing each channel on a seperately
    plot1 = plt.figure(j*5+i)
    #Using ravel instead of flatten, since it's more memory efficient
    _ = plt.hist(image[:, :, i].ravel(),  color = color_dictionary[i], alpha = 1)
    _ = plt.xlabel('Intensity Values')
    _ = plt.ylabel('Count')

  plt.show()






```

Standardizing the images

```python
for current_image in five_images: #Iterating through the image list
  for i in range(3): #Do it for every channel
    current_mean = np.mean(current_image[:,:,i])
    current_std = np.std(current_image[:,:,i])
 
    

    #Subtracting the mean and dividing with the deviation
    current_image[:,:,i] = np.subtract(current_image[:,:,i],current_mean)
    current_image[:,:,i] = np.divide(current_image[:,:,i],current_std)

    #Normalizing the data

    current_min = np.min(current_image[:,:,i])
    current_max = np.max(current_image[:,:,i])


    current_image[:,:,i] = np.subtract(current_image[:,:,i],current_min)
    current_image[:,:,i] = np.divide(current_image[:,:,i],current_max-current_min)

    current_image[:,:,i] = np.subtract(current_image[:,:,i],0.5)
    
    #Subtracting 0.5 to have a mean of 0
	  
```

```python
#Visualize the changed images
for j in range(5):

  

  for i in range(3): #Visualizing each channel a seperately
    plot1 = plt.figure(j*5+i)
    #Using ravel instead of flatten, since it's more memory efficient
    _ = plt.hist(five_images[j][:, :, i].ravel(),  color = color_dictionary[i], alpha = 1)
    _ = plt.xlabel('Intensity Values for image number ' + str(j))
    _ = plt.ylabel('Count')

  plt.show()
```

Working with sound files

```python

#import the pyplot and wavfile modules 

import matplotlib.pyplot as plot

from scipy.io import wavfile

 

# Read the wav file

samplingFrequency, signalData = wavfile.read('maybe-next-time-huh.wav')

 

# Plot the signal read from wav file

plot.subplot(211)

plot.title('Spectrogram of first wav file')

 

plot.plot(signalData)

plot.xlabel('Sample')

plot.ylabel('Amplitude')


plot.show()
```

```python
#Visualizing the not yet normalized values
for i in range(2):
    ploti = plt.figure(i)
    _ = plt.hist(signalData[:,i],  color = color_dictionary[i], alpha = 1)
    _ = plt.xlabel('Intensity Values for channel '+str(i))
    _ = plt.ylabel('Count')
    
    plt.show()

```

```python
#Normalizing the values
new_signal_data = signalData.copy().astype('float32')
for i in range(2): #Do it for every channel
    current_mean = np.mean(new_signal_data[:,i])
    current_std = np.std(new_signal_data[:,i])
 
    

    #Subtracting the mean and dividing with the deviation
    new_signal_data[:,i] = np.subtract(new_signal_data[:,i],current_mean)
    new_signal_data[:,i] = np.divide(new_signal_data[:,i],current_std)

    #Normalizing the data
    current_min = np.min(new_signal_data[:,i])
    current_max = np.max(new_signal_data[:,i])
    new_signal_data[:,i] = np.subtract(new_signal_data[:,i],current_min)
    new_signal_data[:,i] = np.divide(new_signal_data[:,i],current_max-current_min)

    #Subtracting 0.5 to have a mean of 0
    new_signal_data[:,i] = np.subtract(new_signal_data[:,i],0.5)

    
	  
```

```python
#Visualizing the normalized values
for i in range(2):
    ploti = plt.figure(i)
    _ = plt.hist(new_signal_data[:,i],  color = color_dictionary[i], alpha = 1)
    _ = plt.xlabel('Intensity Values for channel '+str(i))
    _ = plt.ylabel('Count')
 
    plt.show()
```
